```python
import pytesseract
from PIL import Image
import requests
import glob
import os
```

```python
#lst = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10']
lst = list(range(1, 1000))
lst = [str(x).zfill(8) for x in lst]
```

```python
urls = ['http://www.gunnerkrigg.com//comics/' + x + '.jpg' for x in lst]
#urls = glob.glob('http://www.gunnerkrigg.com//comics/*.jpg')
```

```python
i = 1
for url in urls:
    if i >= 500:
        break
    f = open(f'{i}.jpg', 'wb')
    f.write(requests.get(url).content)
    f.close()
    i += 1
```

```python
output = []
```

```python
def ocr_extraction(filename):
    text = pytesseract.image_to_string(Image.open(filename))
    return text
```

```python
files = glob.glob('*.jpg')
```
